```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("../img/mr.jpeg")
plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(img[..., ::-1])
plt.show()
```

```python
imgGray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
filter_size=7
pad_size = filter_size // 2
padded_image = np.pad(imgGray, pad_size, mode='constant', constant_values=0)
filtered_image = np.zeros_like(imgGray)

for i in range(imgGray.shape[0]):
    for j in range(imgGray.shape[1]):
        neighborhood = padded_image[i:i + filter_size, j:j + filter_size]
        median_value = np.median(neighborhood)
        filtered_image[i, j] = median_value
plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(filtered_image[..., ::-1],cmap="grey")
plt.show()
```

```python
final = cv2.medianBlur(imgGray, 7)

plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(final[..., ::-1],cmap="grey")
plt.show()
```
