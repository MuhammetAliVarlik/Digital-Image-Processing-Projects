```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("../img/qrCode.jpg")
plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(img[..., ::-1])
plt.show()
```

```python
qcd = cv2.QRCodeDetector()

retval, decoded_info, points, straight_qrcode = qcd.detectAndDecodeMulti(img)
```

```python
newImg=img.copy()
newImg = cv2.polylines(newImg, points.astype(int), True, (0, 255, 0), 3)

for s, p in zip(decoded_info, points):
    newImg = cv2.putText(newImg, s, p[0].astype(int),
                      cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0, 0, 255), 1, cv2.LINE_AA)
plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(newImg[..., ::-1])
plt.show()
```
