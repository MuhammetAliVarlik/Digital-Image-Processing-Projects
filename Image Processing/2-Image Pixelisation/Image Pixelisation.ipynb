```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("../img/sunflowers.jpg")
```

```python
plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(img[..., ::-1])
plt.show()
```

```python
h,w=img.shape[:2]
print(w,h)
```

```python
pixel_size=10
new_w,new_h=(w//pixel_size,h//pixel_size)
print(new_w,new_h)
```

```python
img_temp=cv2.resize(img,(new_w,new_h),interpolation=cv2.INTER_LINEAR)

plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(img_temp[..., ::-1])
plt.show()
```

```python
newImg=cv2.resize(img_temp,(w,h),interpolation=cv2.INTER_NEAREST)


plt.figure(figsize=(14, 5))
# BGR to RGB [..., ::-1]
plt.imshow(newImg[..., ::-1])
plt.show()
```

```python
def pixelisation(img,pixel_size):
    h,w=img.shape[:2]
    new_w,new_h=(w//pixel_size,h//pixel_size)
    img_temp=cv2.resize(img,(new_w,new_h),interpolation=cv2.INTER_LINEAR)
    newImg=cv2.resize(img_temp,(w,h),interpolation=cv2.INTER_NEAREST)
    return newImg
```

```python
dog=cv2.imread("../img/dog.jpg")
pixelDog=pixelisation(dog,30)
plt.figure(figsize=(14, 5))
plt.subplot(1,2,1)
plt.imshow(dog[...,::-1])
plt.subplot(1,2,2)
plt.imshow(pixelDog[...,::-1])
```

```python
aybu=cv2.imread("../img/AYBU_Logo.png")
pixelaybu=pixelisation(aybu,20)
plt.figure(figsize=(14, 5))
plt.subplot(1,2,1)
plt.imshow(aybu[...,::-1])
plt.subplot(1,2,2)
plt.imshow(pixelaybu[...,::-1])
```
